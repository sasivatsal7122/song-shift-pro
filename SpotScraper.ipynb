```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains

from selenium import webdriver
import time
```

```python
spotUrl = "https://open.spotify.com/playlist/0ABPizHAgjCEc0h44gd9ea"
```

```python
def getDriver():
    options=webdriver.ChromeOptions()
    options.add_argument(r"--user-data-dir=/home/satyasasivatsal/.config/google-chrome")
    options.add_argument(r'--profile-directory=Default')
    options.add_experimental_option('excludeSwitches', ['enable-logging'])
    driver = webdriver.Chrome(options=options)
    return driver
```

```python
driver = getDriver()
```

```python
driver.get(spotUrl)
```

```python
songInfo = list()
```

```python
playlist_container = driver.find_element(By.XPATH, "//div[@data-testid='playlist-tracklist']")
```

```python
while 1:
    if len(songInfo)>=590:
        break
    songRows = driver.find_elements(By.CSS_SELECTOR,"div.h4HgbO_Uu1JYg5UGANeQ.wTUruPetkKdWAR1dd6w4")
    for songRow in songRows:
        songTitle = songRow.find_element(By.CSS_SELECTOR,"div.fZDcWX.t_yrXoUO3qGsJS4Y6iXX").text
        songArtist = songRow.find_element(By.CSS_SELECTOR,"span.bDHxRN.rq2VQ5mb9SDAFWbBIUIn").text

        combined = songTitle+" By "+songArtist
        if combined not in songInfo:
            songInfo.append(combined)

    playlist_container.send_keys(Keys.PAGE_DOWN)
    time.sleep(1)  
    
    songRows = driver.find_elements(By.CSS_SELECTOR,"div.h4HgbO_Uu1JYg5UGANeQ.wTUruPetkKdWAR1dd6w4")
    for songRow in songRows:
        songTitle = songRow.find_element(By.CSS_SELECTOR,"div.fZDcWX.t_yrXoUO3qGsJS4Y6iXX").text
        songArtist = songRow.find_element(By.CSS_SELECTOR,"span.bDHxRN.rq2VQ5mb9SDAFWbBIUIn").text

        combined = songTitle+" By "+songArtist
        if combined not in songInfo:
            songInfo.append(combined)

            
```
